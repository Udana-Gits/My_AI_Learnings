```python
import numpy as np

def plot_decision_boundary(model, X, y):
    """
    Plots the decision boundary created by a model predicting on X.
    Adapted from:
    1. CS231n - https://cs231n.github.io/neural-networks-case-study/
    2. Made With ML - https://github.com/GokuMohandas/MadeWithML
    """
    # Define the axis boundaries of the plot and create a meshgrid
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))

    # Create input data from the meshgrid
    x_in = np.c_[xx.ravel(), yy.ravel()]

    # Make predictions
    y_pred = model.predict(x_in)

    # Handle multiclass vs binary classification
    if len(y_pred[0]) > 1:
        print("Doing multiclass classification ...")
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Doing binary classification ...")
        y_pred = np.round(y_pred).reshape(xx.shape)

    # Plot decision boundary and data points
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("Decision Boundary")
    plt.show()
```

```python
# View the predictions of the model with relu and without sigmoid activations
plot_decision_boundary(model_5, X, y)
```

```python
# View the predictions of the model with relu and sigmoid activations
plot_decision_boundary(model_6, X, y)
```

```python
# Set random seed
tf.random.set_seed(42)
# Create a model (same as model_8)
model_9 = tf.keras.Sequential([
tf.keras.layers.Dense(4, activation="relu"),
tf.keras.layers.Dense(4, activation="relu"),
tf.keras.layers.Dense(1, activation="sigmoid")
])
# Compile the model
model_9.compile(loss="binary_crossentropy", # we can use strings here too
optimizer="Adam", # same as tf.keras.optimizers.Adam() with default settings
metrics=["accuracy"])
```

```python
# Create a learning rate scheduler callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10 ** (epoch/20)) # traverse a set of learning rate
```

```python
# Fit the model (passing the lr_scheduler callback)
history_model9 = model_9.fit(X_train,
y_train,
epochs=100,
callbacks=[lr_scheduler], verbose=0)
```

```python
# Checkout the history
pd.DataFrame(history_model9.history).plot(figsize=(5,3.5), xlabel="epochs");
```

```python
# Plot the learning rate versus the loss
lrs = 1e-4 * (10 ** (np.arange(100)/20))
plt.figure(figsize=(5,3.5))
plt. semilogx(lrs, history_model9.history["loss"]) # we want the x-axis (learning rate) to be log scale
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning rate vs. loss");
```
